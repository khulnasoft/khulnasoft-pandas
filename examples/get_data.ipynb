```python
from khulnasoft_pandas.data import get_data
```

Lets pull some data from our two demo servers into a pandas dataframe.

```python
hosts = ['london.my-khulnasoft.com', 'newyork.my-khulnasoft.com']
charts = ['system.cpu', 'system.load']
after = -60
before = 0

df = get_data(hosts, charts, after, before)
print(df.shape)
df.head()
```

Lets have a look at some timings in pulling this data.

```python
%%timeit

df = get_data(hosts, charts, after, before)
```

If we wanted to pull data for all charts we can do something like this.

```python
df = get_data('london.my-khulnasoft.com', ['all'], after, before)
print(df.shape)
df.head()
```

Similar to above but lets add some flags to throw away columns we are probably not interested in.

```python
after = -60*5
before = 0

df = get_data('london.my-khulnasoft.com', ['all'], after, before, numeric_only=True, nunique_thold=0.05)
print(df.shape)
df.head()
```
